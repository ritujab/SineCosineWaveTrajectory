```python
import numpy as np
import matplotlib.pyplot as plt

# Define the first trajectory as a sine wave
def trajectory1(t):
    return np.sin(t)

# Define the second trajectory as a cosine wave
def trajectory2(t):
    return np.cos(t)

# Define the time range over which to evaluate the trajectories
t = np.linspace(0, 2*np.pi, 100)

# Generating the points along each trajectory
points1 = np.array([t, trajectory1(t)]).T
points2 = np.array([t, trajectory2(t)]).T

# Finally plotting the trajectories
plt.plot(points1[:, 0], points1[:, 1], c='purple')
plt.plot(points2[:, 0], points2[:, 1], c='orange')
plt.show()
```
